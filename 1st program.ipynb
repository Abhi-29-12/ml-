```python
import statistics
var = [10,20,30,40,20]
a = statistics.mean(var)
b = statistics.median(var)
c = statistics.mode(var)
d = statistics.stdev(var)

print("Mean=",a)
print("Median=",b)
print("Mode=",c)
print("STDEV",d)
```

```python
import pandas as pd
df = pd.read_excel('STAT.xlsx')
a = df ['Marks'].mean()
b = df ['Marks'].median()
c = df ['Marks'].mode()
d = df ['Marks'].std()
e = df ['Marks'].max()
f = df ['Marks'].min()
print("Mean=",a)
print("Median=",b)
print("Mode=",c)
print("STD=",d)
print("Max=",e)
print("Min=",f)
```

```python
import numpy as np
val = [50,20,70,40,50,60,80]
a = np.mean(val)
b = np.median(val)
c = np.std(val)
d = np.max(val)
e = np.min(val)
print("Mean=",a)
print("Median=",b)
print("STD=",c)
print("Max=",d)
print("Min=",e)
```

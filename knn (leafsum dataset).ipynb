```python
# KNN Leaf Health Prediction using Excel Data

# Step 1: Import Libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Step 2: Read Excel file
# Ensure your file is named "LEAF_DATA.xlsx" in the same folder as this script
data = pd.read_excel("leafsum.xlsx")

# Step 3: Display dataset
print("\nDataset:\n")
print(data)

# Step 4: Separate features and Label
X = data[['Moisture', 'Temperature']] # Input features
y = data['Class']                    # Target output

# Step 5: Create KNN Model (K=3, Euclidean distance)
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')

# Step 6: Train the model
knn.fit(X, y)

# Step 7: New data (Test data)
# Moisture 75, Temperature 27
new_leaf = [[75, 27]]

# Step 8: Predict result
prediction = knn.predict(new_leaf)

# Step 9: Display result
print("\nNew Leaf Data (Moisture 75, Temperature 27)")
print("Predicted Class:", prediction[0])

# --- Step 11: Visualization ---

# Separate classes for plotting
# .str.strip() removes hidden spaces that might prevent points from showing
healthy = data[data['Class'].str.strip() == 'Healthy']
disease = data[data['Class'].str.strip() == 'Diseased']

# Plot Healthy (Green)
plt.scatter(healthy['Moisture'], 
            healthy['Temperature'], 
            color='green', 
            label='Healthy', 
            s=100)

# Plot Disease (Red)
plt.scatter(disease['Moisture'], 
            disease['Temperature'], 
            color='red', 
            label='Diseased', 
            s=100)

# Plot New Leaf (Blue Star)
plt.scatter(75, 27, 
            color='blue', 
            marker='*', 
            s=250, 
            label='New Leaf')

# Labels and Title
plt.xlabel("Moisture")
plt.ylabel("Temperature")
plt.title("KNN Classification of Leaf Health")

# Legend and Show graph
plt.legend()
plt.show()
```

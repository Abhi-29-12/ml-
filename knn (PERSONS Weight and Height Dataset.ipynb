```python
# KNN Body Type Prediction using Excel Data

# Step 1: Import Libraries
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Step 2: Read Excel file
# Ensure your file is named "BODY_DATA.xlsx"
data = pd.read_excel("sum.xlsx")

# Step 3: Display dataset
print("\nDataset:\n")
print(data)

# Step 4: Separate features and Label
X = data[['Weight', 'Height']] # Input features
y = data['Class']              # Target output

# Step 5: Create KNN Model (K=3, Euclidean distance)
knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')

# Step 6: Train the model
knn.fit(X, y)

# Step 7: New student data (Test data)
# Weight 57kg, Height 170cm
new_data = [[57, 170]]

# Step 8: Predict result
prediction = knn.predict(new_data)

# Step 9: Display result
print("\nNew Data (Weight 57, Height 170)")
print("Predicted Class:", prediction[0])

# --- Visualization ---

# Separate classes for plotting
normal = data[data['Class'] == 'Normal']
underweight = data[data['Class'] == 'Underweight']

# Plot Normal (Green)
plt.scatter(normal['Weight'], 
            normal['Height'], 
            color='green', 
            label='Normal', 
            s=100)

# Plot Underweight (Red)
plt.scatter(underweight['Weight'], 
            underweight['Height'], 
            color='red', 
            label='Underweight', 
            s=100)

# Plot New Data (Blue Star)
plt.scatter(57, 170, 
            color='blue', 
            marker='*', 
            s=250, 
            label='New Data')

# Labels and Title
plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.title("KNN Classification of Body Type")

# Legend and Show graph
plt.legend()
plt.show()
```

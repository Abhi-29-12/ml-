```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB

# 1. Dataset
data = pd.DataFrame({
    'Color': ['Red','Red','Red','Yellow','Yellow','Yellow',
                'Yellow','Yellow','Red','Red'],
    'Type':  ['Sports','Sports','Sports','Sports','Sports','SUV',
                'SUV','SUV','SUV','Sports'],
    'Origin': ['Domestic','Domestic','Domestic','Domestic','Imported','Imported','Imported','Domestic','Imported','Imported'],
    'Stolen': ['Yes','No','Yes','No','Yes','No','Yes','No','No','Yes']
})

# 2. Split Features (X) and Target (z)
# We group all features into one DataFrame X
X = data[['Color', 'Type', 'Origin']].copy()
z = data['Stolen'].copy()

# 3. Encoding
# We need an encoder for each categorical column
le_color = LabelEncoder()
le_type = LabelEncoder()
le_origin = LabelEncoder()
le_stolen = LabelEncoder()

X['Color'] = le_color.fit_transform(X['Color'])
X['Type'] = le_type.fit_transform(X['Type'])
X['Origin'] = le_origin.fit_transform(X['Origin'])
z = le_stolen.fit_transform(z)

# 4. Model Training
# .fit() now receives exactly 2 arguments: the feature matrix and the target vector
model = GaussianNB()
model.fit(X, z)

# 5. Prediction
# Create a single DataFrame containing all features in the correct order
new_data = pd.DataFrame({
    'Color': le_color.transform(['Red']),
    'Type': le_type.transform(['SUV']),
    'Origin': le_origin.transform(['Domestic'])
})

prediction = model.predict(new_data)
result = le_stolen.inverse_transform(prediction)

print("--- Result ---")
print(f"Prediction for Red, SUV, Domestic: {result[0]}")
```
